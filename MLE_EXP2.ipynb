```python
import seaborn as sns
penguins_df = sns.load_dataset('penguins')
df = penguins_df
display(df.head())
```

```python
plt.hist(df['bill_length_mm'].dropna())
plt.title("Histogram - Bill Length (Penguins)")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(df['bill_length_mm'], df['bill_depth_mm'])
plt.title("Scatter Plot - Bill Dimensions (Penguins)")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
plt.imshow(corr)
plt.colorbar()
plt.title("Heatmap - Correlation Matrix (Penguins)")
plt.show()
```

```python
plt.plot(df['flipper_length_mm'])
plt.title("Line Plot - Flipper Length (Penguins)")
plt.xlabel("Sample Index")
plt.ylabel("Flipper Length (mm)")
plt.show()
```

```python
species_counts = df['target'].value_counts()
labels = ['Setosa', 'Versicolor', 'Virginica']
plt.figure(figsize=(8, 8))
plt.pie(species_counts, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Iris Species')
plt.axis('equal')
plt.show()
```

```python
species_counts_penguins = df['species'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(species_counts_penguins, labels=species_counts_penguins.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Penguin Species')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='species', data=df)
plt.title('Distribution of Penguin Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```
