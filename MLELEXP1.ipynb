```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Customers Data.csv')

df = df.drop_duplicates()

df['email'] = df['email'].astype(str)
df.loc[~df['email'].str.match(r"[^@]+@[^@]+\.[^@]+"), 'email'] = np.nan
df['email'] = df['email'].fillna('abc@gmail.com')
df['name'] = df['name'].fillna('Tejas')
df.to_csv('employee_cleaned2.csv', index=False)

print("Data cleaning completed successfully.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='city', order = df['city'].value_counts().index, palette='viridis')
plt.title('Distribution of Customers by City (Cleaned Data)')
plt.xlabel('Number of Customers')
plt.ylabel('City')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', bins=10, kde=True, palette='coolwarm')
plt.title('Distribution of Customers by Age (Cleaned Data)')
plt.xlabel('Age')
plt.ylabel('Number of Customers')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_sorted_by_age = df.sort_values(by='age').reset_index(drop=True)

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_sorted_by_age, x=df_sorted_by_age.index, y='age', marker='o', linestyle='-', color='purple')
plt.title('Sorted Distribution of Customers by Age (Cleaned Data)')
plt.xlabel('Customer Index (Sorted by Age)')
plt.ylabel('Age')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
